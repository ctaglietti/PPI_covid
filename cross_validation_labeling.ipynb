```python
import pandas as pd

df=pd.read_csv('alla_scores.csv')

df[268:9134]
```

```python
import random

#generates random numbers that we can use to choose random negatives
nums=random.sample(range(268, 9134), 268)

#makes sure each element is unique 
nums = list(dict.fromkeys(nums))

```

```python
print(len(df['prot']))
```

```python
validation=[]

for i in range(9132):
    
    if i in nums: #adds the negatives
        validation.append(0)
    if i in range(267): #adds the positives (the first 267 proteins in the list)
        validation.append(1)   
    if i not in range(267) and i not in nums: #tags everything else as 2 so you can filter it out
        validation.append(2)
```

```python
print(len(validation))
```

```python
# adds a column to the dataframe with the validation labels
df['validation_set']=validation

df
```

```python
df.to_csv('validation.csv')
```
